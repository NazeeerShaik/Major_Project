```python
from nltk.corpus import stopwords;

sarc_tweets['type'] = 'sarcastic';
gen_tweets['type'] = 'genuine';

import re
from bs4 import BeautifulSoup


def review_to_words(raw_tweets):
    # 1. Remove html tags
    #review_text = BeautifulSoup(raw_review,"lxml").get_text()
    
    # 2. Remove non-letters
    letters_only = re.sub("[^a-zA-Z#]", " ", raw_tweets)
    letters_only = letters_only.replace('sarcasm','')
    
    # 3. split individual words by converting in lowercase
    words = letters_only.lower().split()
    
    # 4. converting stop words to set for fast searching
    stops = set(stopwords.words("english"))

    
    # 5. removing stop words
    meaningful_words = [w for w in words if not w in stops]
    
    # 6. joining the words back into one string separated by space and return
    return(" ".join(meaningful_words))
```

```python
num_of_tweetss = sarc_tweets[0].size
# list to hold clean tweets
clean_tweets_sarcastic = []

for i in range(0, num_of_tweetss):
    #clean tweets
    clean_tweets_sarcastic.append(review_to_words(sarc_tweets[0][i]))

num_of_tweetsg = gen_tweets[0].size
clean_tweets_genuine = [];

for i in range(0, num_of_tweetsg):
    clean_tweets_genuine.append(review_to_words(gen_tweets[0][i]))

    
sarcastic_tweets_parsed = [n.replace('sarcasm','') for n in clean_tweets_sarcastic]    
```

```python
from nltk.stem.porter import *
ps = PorterStemmer()
sarcastic_tweets_stemmed = [];

for tweet in range(0,len(sarcastic_tweets_parsed)):
    newTweet = []
    for word in sarcastic_tweets_parsed[tweet].split(" "):
        newTweet.append(ps.stem(word));
    sarcastic_tweets_stemmed.append(" ".join(newTweet));
```

```python
from nltk.stem.porter import *
ps = PorterStemmer()
genuine_tweets_stemmed = [];

for tweet in range(0,len(clean_tweets_genuine)):
    newTweet = []
    for word in clean_tweets_genuine[tweet].split(" "):
        newTweet.append(ps.stem(word));
    genuine_tweets_stemmed.append(" ".join(newTweet));
```

```python
sarcastic_tweets = sarc_tweets
genuine_tweets = gen_tweets
sarc_tweets = pd.DataFrame(sarcastic_tweets_stemmed);
gen_tweets = pd.DataFrame(genuine_tweets_stemmed);
```

```python
import langid as li;

li.NORM_PROBS = True;
li.NORM_PROBS == True;

li.classify(sarc_tweets[0][4])
sarc_tweets["English"] = sarcastic_tweets[0].map(lambda x: 1 if li.classify(x)[0]=='en' else 0);
gen_tweets["English"] = genuine_tweets[0].map(lambda x: 1 if li.classify(x)[0]=='en' else 0);
```

### Feature engineering

```python
#ToUser
sarc_tweets["ToUser"] = sarc_tweets[0].map(lambda x: 1 if "@" in x else 0)
gen_tweets["ToUser"] = gen_tweets[0].map(lambda x: 1 if "@" in x else 0);

#HashTags presence
sarc_tweets["Hashtags"] = sarc_tweets[0].map(lambda x: 1 if "#" in x else 0)
gen_tweets["Hashtags"] = gen_tweets[0].map(lambda x: 1 if "#" in x else 0);

#all caps count
sarc_tweets['AllCapsCount'] = [sum([k.isupper() for k in n.split(" ")]) for n in sarc_tweets[0]]
gen_tweets['AllCapsCount']= [sum([k.isupper() for k in n.split(" ")]) for n in gen_tweets[0]]
```

```python
sarc_tweets["type"]="sarcastic";
gen_tweets["type"]="genuine";
```

```python
sarcastic_tweets_english = sarc_tweets[sarc_tweets.English==1]
genuine_tweets_english = gen_tweets[gen_tweets.English == 1]
```

```python
sarcastic_tweets_english.to_csv("processing/sarcastic_tweets_eng.csv");
genuine_tweets_english.to_csv("processing/genuine_tweets_eng.csv");
```

```python
master = sarcastic_tweets_english.append(genuine_tweets_english)
master.to_csv('processing/all_tweets_df.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = "word", \
                            tokenizer = None,   \
                            preprocessor = None, \
                            stop_words = None)

count_vec = CountVectorizer(max_features = 5000)
vector = count_vec.fit_transform(subset[0])
varr = vector.toarray();
count_vec.get_feature_names()
```

### SVC

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(unigrams, subset['type'], test_size=0.4, random_state=0)
```

```python
SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape=None, degree=3, gamma='auto', kernel='rbf',
    max_iter=-1, probability=False, random_state=None, shrinking=True,
    tol=0.001, verbose=False)
```

```python
clf.score(X_test,y_test)
```

```python
svc_predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import metrics
svc_preds_numeric = [1 if x=='sarcastic' else 0 for x in svc_predictions]
y_test_numeric = [1 if x=='sarcastic' else 0 for x in y_test]
fpr_svc, tpr_svc, thresh_svc = metrics.roc_curve(y_test_numeric, svc_preds_numeric)
```

```python
accuracy_score(y_test, svc_predictions)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
lw = 2
plt.plot(fpr_svc, tpr_svc, color='blue',lw=lw, label='ROC curve',)
plt.title("SVM ROC")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```
